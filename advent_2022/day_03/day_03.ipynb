# Day 3 - Rucksack Reorganization

## Part 1

Each line is a list of items in an elf's rucksack (split into two compartments).
Need to find the item that is duplicated (one in each compartment = two in the rucksack).

```python
with open("input.txt") as f:
    rucksacks = [x.strip() for x in f.readlines()]
    
rucksacks[0:5]
```

```python
# Define function to split the string into two pieces
def ruck_split(rucksack):
    
    half_length = len(rucksack)//2
    
    comp_1 = set(rucksack[0:half_length])
    
    comp_2 = set(rucksack[half_length:len(rucksack)])
    
    inter = comp_1.intersection(comp_2)
    
    return(list(inter).pop())
```

```python
dup_items = []

for ruck in rucksacks:
    dup_items.append(ruck_split(ruck))
    
dup_items[0:5]
```

```python
# Define function to cast the letters to their priorities
def priorities(letter_list):
    
    scores = []
    
    for letter in letter_list:
        if letter.isupper():
            scores.append(ord(letter)-38)
        else:
            scores.append(ord(letter)-96)
            
    return(sum(scores))
```

```python
priorities(dup_items)
```

## Part 2

Elves are in groups of 3, need to find common letter between the three for scoring.

```python
rucksacks[0:5]
```

```python
#define function to compare 3 elves' items in sets
def elf_3_comp(rucks_3):
    
    rucks = set(rucks_3[0]).intersection(set(rucks_3[1]).intersection(set(rucks_3[2])))
    
    return(list(rucks).pop())
```

```python
#Going to have to loop through everything

elf_list = []
elf_count = 0
items_list = []

for i in range(0, len(rucksacks)):
    
    if elf_count < 2:
        elf_list.append(rucksacks[i])
        elf_count += 1
    else:
        elf_list.append(rucksacks[i])
        items_list.append(elf_3_comp(elf_list))        
        elf_list = []
        elf_count = 0
```

```python
priorities(items_list)
```
