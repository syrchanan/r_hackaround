# Day 1 - Calorie Counting

```python
import numpy as np
import pandas as pd
```

## Part 1

Elves carrying food, need to subtotal and get the amount of the elf carrying the most

```python
with open("day_01_input.txt") as f:
    elves = [x.strip() for x in f.readlines()]
```

Now that the whitespace is stripped out, can cast to int and subtotal each

```python
'#'.join(elves)
```

```python
'#'.join(elves).split('##')
```

```python
[x.split('#') for x in '#'.join(elves).split('##')]
```

```python
[[int(y) for y in x.split('#')] for x in '#'.join(elves).split('##')]
```

```python
elves_calories = [sum(z) for z in [[int(y) for y in x.split('#')] for x in '#'.join(elves).split('##')]]

elves_calories
```

Now we have a list of all the calories each elf is carrying, need to get max

```python
f"The elf carrying the most calories is carrying {int(max(elves_calories))} calories"
```

## Part 2

Need to subtotal the top 3 elves

```python
elves_calories.sort(reverse = True)
```

```python
f"The three elves carrying the most calories are each carrying {elves_calories[0:3]}, respectively. Together, they carry a total of {sum(elves_calories[0:3])} calories."
```
